```python
def create_and_clean_dataframe(sheet: str = None, path: str = None) -> pd.DataFrame():
    '''
    Takes in the sheet name of the excel file and reads the file according
    to the constraints applied to the specific sheets. Also, it cleans the
    dataframe once the excel information has been converted to a dataframe.
    Then, it merges the Section column with the Batches and returns the df.
    '''

    if sheet not in globals()['sheet_names']:
        raise Exception('Error: Sheet not found')
    if sheet == 'S&H':
        try:
            df = pd.read_excel(path, sheet_name = sheet, skiprows = 2, usecols = [0, 2, 4, 5], skipfooter = 58)
        except Exception:
            print("Error: an exception occurred while opening the document")
    else:
        try:
            df = pd.read_excel(path, sheet_name = sheet, skiprows = 2, usecols = [0, 2, 4, 5])
        except Exception:
            print("Error: an exception occurred while opening the document")

    df = df.dropna(subset = ['Course', 'Section', 'Course Instructor'], how = 'all').reset_index(drop = True)
    df['S.#'] = df['S.#'].str.cat(df['Section'], sep = ' , ')

    return df.drop(['Section'], axis = 1)


def map_timetable_to_dataframe(timetable: pd.DataFrame() = None, timetable_list: list = None) -> pd.DataFrame():
    '''
    Takes in the empty timetable dataframe and the generated timetable
    list as its parameters. Then, it maps the values in the generated
    timetable index by index to the pandas dataframe corresponding to
    those indexes. Returns the newly created timetable dataframe.
    '''

    for i in range(len(globals()['room'])):
        for j in range(len(globals()['time'])):
            timetable.iloc[i,j] = ' '.join(timetable_list[i][j])

    return timetable

def break_chromosome(chromosome: list = None, mid: int = 2) -> list:
    '''
    Takes in a chromosome (timetable) and a mid index value as parameters
    and breaks the chromosome into two equal pieces, then returns those
    two pieces.
    '''
    return chromosome[:mid], chromosome[mid:]
```

```python
def binary_encode_chromosome(chromosome_info):
    '''
    Binary encodes the given chromosome information.

    Parameters:
    chromosome_info (list): List containing chromosome information in the order:
    [Course, Theory/Lab, Section, Section-Strength, Professor, First-lecture-day,
    First-lecture-timeslot, First-lecture-room, First-lecture-room-size, Second-
    lecture-day, Second-lecture-timeslot, Second-lecture-room, Second-lecture-
    room-size]

    '''
    binary_encoded_chromosome = []

    for item in chromosome_info:
        binary_value = bin(item)[2:]  # Convert to binary, remove '0b' prefix
        binary_encoded_chromosome.append(binary_value)

    return binary_encoded_chromosome
```

```python
def fitness(chromosome: list = None) -> int:
    '''
    Takes in the chromosome (generated timetable) as the parameter and checks
    all the hard and soft constraints that are applied to the timetable. The
    constraints are as follows:
          i. Teacher can't have two classes at the same time
         ii. Student can't have two classes at the same time
        iii. Teacher can't have more than 3 classes in a day
         iv. Student can't have more than 4 classes in a day
    These constraints are individually checked in this function and for each
    constraint that has been cleared, points are added to the score system.
    Hard constrains (first 2) get more points than the the soft constrains.
    Also, to not get the function to return the same score due to repitition,
    score deduction has also been added corresponding to the number of times
    a constraint is violated. Then it returns the final score.
    '''

    score = 0

    # Checking if teacher has two classes at the same time'
    count = 0
    flag = False
    for i in range(len(globals()['room'])):
        for j in range(len(globals()['time'])):
            temp = chromosome[i][j][2]
            for k in range(i + 1, len(globals()['room'])):
                if temp == chromosome[k][j][2]:
                    count = count + 1
                    flag = True
    if not flag:
        score = score + 36
    else:
        score = score + (36 - (count * 6))

    # Checking if student has two classes at the same time
    count = 0
    flag = False
    for i in range(len(globals()['room'])):
        for j in range(len(globals()['time'])):
            temp = chromosome[i][j][0]
            for k in range(i + 1, len(globals()['room'])):
                if temp == chromosome[k][j][0]:
                    count = count + 1
                    flag = True
    if not flag:
        score = score + 36
    else:
        score = score + (36 - (count * 6))

    # Checking if teacher has more than 3 classes in a day
    count = 0
    check_teacher = [chromosome[j][i][2] for i in range(len(globals()['time'])) for j in range(len(globals()['room']))]
    c = Counter(check_teacher)
    for key, value in c.items():
        if value > 3:
            count = count + 1
    if count == 0:
        score = score + 14
    else:
        score = score + (14 - count)

    # Checking if student has more than 4 classes in a day
    count = 0
    check_student = [chromosome[j][i][0] for i in range(len(globals()['time'])) for j in range(len(globals()['room']))]
    c = Counter(check_student)
    for key, value in c.items():
        if value > 4:
            count = count + 1
    if count == 0:
        score = score + 14
    else:
        score = score + (14 - count)

    scoree = 0
    '''
    #Checking if classrooms are big enough to accommodate the section
    for i in range(len(globals()['room'])):
        for j in range(len(globals()['time'])):
            section_size = chromosome[i][j][1]
            room_capacity = globals()['room'][i][1]  # Assuming room list contains room capacity information
            if section_size > room_capacity:
                scoree -= 50  # Penalize heavily for this constraint violation

    # The same section cannot be assigned to two different rooms at the same time
    for i in range(len(globals()['room'])):
        for j in range(len(globals()['time'])):
            section = chromosome[i][j][0]
            for k in range(i + 1, len(globals()['room'])):
                if section == chromosome[k][j][0]:
                    scoree -= 50  # Penalize heavily for this constraint violation

    #A room cannot be assigned for two different sections at the same time
    for i in range(len(globals()['room'])):
        for j in range(len(globals()['time'])):
            room = chromosome[i][j][2]
            for k in range(i + 1, len(globals()['room'])):
                if room == chromosome[k][j][2]:
                    scoree -= 50  # Penalize heavily for this constraint violation

    # No professor can teach more than 3 courses
    teacher_courses = [chromosome[i][j][2] for i in range(len(globals()['room'])) for j in range(len(globals()['time']))]
    c = Counter(teacher_courses)
    for key, value in c.items():
        if value > 3:
            scoree -= 100  # Penalize heavily for this constraint violation

    #No section can have more than 5 courses in a semester
    section_courses = [chromosome[i][j][0] for i in range(len(globals()['room'])) for j in range(len(globals()['time']))]
    c = Counter(section_courses)
    for key, value in c.items():
        if value > 5:
            scoree -= 100  # Penalize heavily for this constraint violation

    # 15 minutes breaks allowed between consecutive classes
    for i in range(len(globals()['room'])):
        for j in range(1, len(globals()['time'])):  # Start from index 1 for time
            if chromosome[i][j][0] != chromosome[i][j-1][0]:  # If current section is different from previous
                scoree += 5  # Award points for having a break between classes

    #SOFT CONSTRAINTS
    # All theory classes in the morning, lab sessions in the afternoon
    for i in range(len(globals()['room'])):
        for j in range(len(globals()['time'])):
            session_type = chromosome[i][j]  # Assuming session type info is stored at index 3
            if session_type == "Theory" and j >= len(globals()['time']) / 2:
                scoree -= 20  # Penalize for theory classes in afternoon
            elif session_type == "Lab" and j < len(globals()['time']) / 2:
                scoree -= 20  # Penalize for lab sessions in morning

    # Minimize floor traversals for teachers and students
    teacher_floors = [chromosome[i][j] for i in range(len(globals()['room'])) for j in range(len(globals()['time']))]
    student_floors = [chromosome[i][j] for i in range(len(globals()['room'])) for j in range(len(globals()['time']))]
    teacher_floors_count = 0
    student_floors_count = 0
    scoree += max(0, 10 - teacher_floors_count)  # Award points for teacher floor consistency
    scoree += max(0, 10 - student_floors_count)  # Award points for student floor consistency

    # A class should be held in the same classroom across the whole week
    class_in_same_room = set()
    for i in range(len(globals()['room'])):
        room_schedule = [chromosome[i][j] for j in range(len(globals()['time']))]
        if room_schedule == 1:  # If the room is used consistently throughout the week
            class_in_same_room.add(room_schedule[0])
    scoree += len(class_in_same_room) * 5  # Award points for each class held in the same room across the week

    # Teachers prefer longer blocks of continuous teaching time
    teacher_blocks = []
    for i in range(len(globals()['room'])):
        room_schedule = [chromosome[i][j] for j in range(len(globals()['time']))]
        teacher_blocks.append(room_schedule)  # Count the number of different courses in a day for each teacher
    scoree -= 5  # Penalize for shorter continuous teaching blocks
    '''
    return score


# This is a very simple selection function. There are many selection functions
# e.g Roulette Wheel Selection, Rank Selection, Steady State Selection, Tournament Selection
# Elitist Selection, Boltzmann Selection, to name a few. I've used a simple straightforward
# top fitness selection method, whhere I select the top 4 based on the fitness score as parents
# from the population (a very naive approach, I know.. but still works... I didn't have
# the time to implement the other selection methods). In the future, I hope to implement them all.
def selection(population: list = None) -> list:
    '''
    Takes in the entire population of generated timetables. This function
    generated another list that has the corresponding fitness scores to the
    individual chromosomes at its indexes. Calculates the best four relative
    to their fitness scores, and fetches their index. We use those indexes
    to return the best two as parents.
    '''


    population_fitness = [fitness(chromosome) for chromosome in population]

#     max_index = population_fitness.index(max(population_fitness))
#     population_fitness.remove(max(population_fitness))
#     second_max_index = population_fitness.index(max(population_fitness))
#     population_fitness.remove(max(population_fitness))
#     third_max_index = population_fitness.index(max(population_fitness))
#     population_fitness.remove(max(population_fitness))
#     fourth_max_index = population_fitness.index(max(population_fitness))

    top_4 = sorted(range(len(population_fitness)), key=lambda x: population_fitness[x])[-4:]
    top_4.reverse()

    parents = [population[top_4[0]], population[top_4[1]], population[top_4[2]], population[top_4[3]]]

    return parents


def crossover(parents: list = None) -> list:
    '''
    Takes in the two parents as the parameters. This function takes the
    parents and stores them separately. Then, splits those parents into
    two equal parts and co-joins them in a way that child one gets the
    first parents first half, and the second parents second half and
    vice versa. Then, it returns those two childs separately.
    '''

    mid_idx = len(parents[0]) // 2

    p1_1st, p1_2nd = break_chromosome(parents[0], mid_idx)
    p2_1st, p2_2nd = break_chromosome(parents[1], mid_idx)
    p3_1st, p3_2nd = break_chromosome(parents[2], mid_idx)
    p4_1st, p4_2nd = break_chromosome(parents[3], mid_idx)

    child_1 = p1_1st + p2_2nd
    child_2 = p2_1st + p1_2nd
    child_3 = p3_1st + p4_2nd
    child_4 = p4_1st + p3_2nd

    return child_1, child_2, child_3, child_4


def mutation(chromosome: list = None, chance: float = 0.1) -> list:
    '''
    Takes in two parameters. The chromosome and the probability (predefined).
    This function iterates over the chromosome index by index and for each
    index, if the random number generated is higher than the probability set
    by the algorithm, a randomly selected allele is fit into that specific
    index of the chromosome. After the iterations, it returns the chromosome.
    '''

    for i in range(len(globals()['room'])):
        if random.random() < chance:
            chromosome[i] = [random.choice(alleles) for i in range(len(globals()['time']))]

    return chromosome
```

### Driver Function

```python
def genetic_algorithm(population: list = None, generations: int = 10) -> list:
    '''
    Takes in the initial population as its parameter. Then for the predefined
    amount of generations, it creates a new population in every iteration of
    generaion, while improving the overall fitness of each of the chromosomes
    present in the population. When the final generation is produced, which has
    the overall highest amount of fitness score, it selecs the best from among
    that generation and returns that chromosome.
    '''
    for generation in range(generations):

        temp_population = list()

        for i in range(globals()['population_size'] // 4):

            parents = selection(population)

            child_1, child_2, child_3, child_4 = crossover(parents)

            child_1 = mutation(child_1, probability)
            child_2 = mutation(child_2, probability)
            child_3 = mutation(child_3, probability)
            child_4 = mutation(child_4, probability)

            temp_population.extend([child_1, child_2, child_3, child_4])

        # Finding out the chromosome with the least fitness score in the newly created population and
        # saving its index, then inserting the chromosome with the highest fitness score in the old
        # population into that saved index.
        temp_population[population.index(min(population, key = fitness))] = max(population, key = fitness)
        population = temp_population

        if fitness(max(population, key = fitness)) == 100:
            print('Chromosome of maximum possible fitness found!')
            print('Generation: ', generation)
            return max(population, key = fitness)

    final_chromosome = max(population, key=fitness)

    return final_chromosome
```
